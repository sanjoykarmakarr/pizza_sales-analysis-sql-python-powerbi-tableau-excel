### Key Performance Indicator

```python
total_revenue =       df["total_price"].sum()         #sum of total price.
total_pizzas_sold =   df["quantity"].sum()            #sum of total quantity.
total_orders =        df["order_id"].nunique()        #"nunique" is a pandas library function for distinct value count.
avg_order_value =     total_revenue/total_orders      
avg_pizza_per_order = total_pizzas_sold/total_orders

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total pizza sold: ${total_pizzas_sold:,}")
print(f"Total Orders: ${total_orders:,}")
print(f"Avg Order Value: ${avg_order_value:,.2f}")
print(f"Avg pizza Per Order: ${avg_pizza_per_order:,.2f}")
```

#### Daily Trend-Total Orders

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)   #converting order_date to datetime.
df['day_name'] = df['order_date'].dt.day_name()                      #extracting weekday names.
weekday_order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]      #defining weekday order.
orders_by_day = (
    df.groupby('day_name')['order_id']                               #its grouping dayname and orderId together.
      .nunique()                                                     #.NUNIQUE=unique values.
      .reindex(weekday_order)                                        #.REINDEX=for arrange days by monday to sunday,not alphabetical order.
)                                                                    #count unique orders by weekday.
ax = orders_by_day.plot.bar(figsize=(8,5), edgecolor='brown')        #Plot setting.
plt.title("Total Orders by Day of Week")                             #using matplotlib for the title.
plt.xlabel("Day of Week")                                            #here creating x-axis for the days.
plt.ylabel("Number of Orders")                                       #here creating y-axis for order range.
plt.xticks(rotation=360)                                             #here i set dayname position.

for i, val in enumerate(orders_by_day):                              #using for-loop.enumerate for returning both index and eliment at the same time.
    ax.text(i,val,str(val),ha='center',va='bottom',fontweight='bold')#here i=x-position (bar number),val=y-position (top of the bar),
                                                                     #str(val)=text displayed (the number itself).ha/va=horizontal/vertical alignment.
plt.tight_layout()                                                   #plt.tight_layout()=is like adjusting space of eliments,just not to overlap things.
plt.show()                                                           #it renders and opens the graph window so we can see the chart.
```

#### Daily Trend-Total Revenue

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)                       #converting order_date to datetime.
weekday_order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"] #extracting weekday names.
orders_by_day = (
    df.assign(day_name=df['order_date'].dt.day_name())               #df.assign=adding new column"day_name"
      .groupby('day_name')['total_price']                            #its grouping dayname and orderId together.
      .sum()                                                         #adding values together.
      .reindex(weekday_order)                                        #.REINDEX=for arrange days by monday to sunday,not alphabetical order.
)                                                                    #count unique orders by weekday.

ax = orders_by_day.plot.bar(figsize=(8,5), color="purple", edgecolor="black")  #Plot setting.
ax.set_title("Total Revenue by Day of Week")                                   #using matplotlib for the title.
ax.set_xlabel("Day of Week")                                                   #here creating x-axis for the days.
ax.set_ylabel("Total Revenue")                                                 #here creating y-axis for price range.
plt.xticks(rotation=0)                                                         #here i set dayname position.

for i, v in enumerate(orders_by_day):                                #using for-loop,.enumerate for returning both index and eliment at the same time.
    ax.text(i, v + 20, f"{v}", ha="center", fontweight="bold")       #here i=x-position (bar number),v=height of the bar(revenue).
                                                                     #(v+20)=adding+20 moves cause doesn't overlap the text slightly above the bar.
plt.tight_layout()                                                   #plt.tight_layout()=is like adjusting space of eliments,just not to overlap things.
plt.show()                                                           #it renders and opens the graph window so we can see the chart.
```

#### Hourly Trend-Total Orders

```python
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S') #order_time convert to_datetime by pandas than change datetime into h:m:s format.
df['order_hour'] = df['order_time'].dt.hour                            #now extract datetime to only date_hours.
orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique() #no order hrs will observed false and nunique for count of order_id's.
ax = orders_by_hour.plot(kind='bar', figsize=(10,5), color='maroon', edgecolor='black')  #ploting the figures

plt.title("Total Orders by Hour of Day")
plt.xlabel("Hour of Day (24-Hour Format)")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)

ax.bar_label(ax.containers[0], padding=3, fontsize=9, fontweight='bold') #here instead of loop function i use matplotlib to take all bars into a chart.

plt.tight_layout()                                                   #plt.tight_layout()=is like adjusting space of eliments,just not to overlap things.
plt.show()                                                           #it renders and opens the graph window so we can see the chart.
```

#### Monthly Trend-Total Orders

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce') #dayfirst=True:d-m-y,errors='coerce'=convert to null date.
df = df.dropna(subset=['order_date']).set_index('order_date').sort_index()          # Remove invalid dates than set index than sort dates.
orders_by_month = df['order_id'].resample('ME').nunique()       #resample('ME')=group data by Month End,nunique()= count unique order IDs
orders_by_month.index = orders_by_month.index.strftime('%B')    #Convert index to month names,%B → full month,%Y → full year.strftime=string format time

plt.figure(figsize=(10,5))                    # Plot
plt.plot(orders_by_month, marker='o')         #marker='o'- adds dots on each point on the graph
plt.fill_between(orders_by_month.index, orders_by_month.values,alpha=0.6)   #alpha=0.6 - makes it slightly transparent

for x, y in zip(orders_by_month.index, orders_by_month.values):  #zip()-combines two lists ( arrays) together into pairs.X=month,Y=value.
    plt.text(x, y, str(y), ha='center')                          #str(y)= convert y into string(text)

plt.xticks(rotation=0)
plt.title("Total Orders by Month")
plt.tight_layout()
plt.show()
```

#### Percentage of Sales by Catagory

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()  #('pizza_category')=function call,['total_price']=select column.
category_pct = category_sales / category_sales.sum() * 100          #.sum()=Add values to each group.category_pct=Percentage calculation.

plt.figure(figsize=(6,6))              #plt.figure=It creates a new blank canvas (graph area),figsize=(width, height)
colors = plt.get_cmap('tab20').colors  #get_cmap → get colors,tab20 = ready-made palette of 20 professional colors for charts,.colors → get the list.

plt.pie(                               #plt.pie=its get pie chart
    category_pct,                      #category_pct(main data of values)=like [27.5, 22.5, 31.3, 18.7].plt converts values into angle of pie's circle.
    labels=category_pct.index,         #category_pct.index=These are the names of each slice.
    autopct="%1.1f%%",                 #autopct="%1.1f%%"=Shows Automatic Percentage text on each slice,%1.1f =float with 1 decimal,%% = print % symbol.
    startangle=90,                     #startangle=90 =where the pie starts and its like 90° = starts from top ↑.
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'width': 0.4}  #“wedgeprops”=properties of each slice of pie.edgecolor= balck border of the pie.
)

plt.title("Percentage of Sales by Pizza Category")
plt.show()
```

#### Percentage Sales by Pizza Size and Category

```python
sales_pivot = df.pivot_table(     #total sales converted into pivot table.
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',                #aggfunc=aggregation function.
    fill_value=0                  #fill_value=0 =It does not shows Null value,it showing NULL values to '0'.
)

sales_pct = sales_pivot / sales_pivot.sum().sum() * 100  #each category sale divided by sum of total category sale and *100 mean Convert into percentage

plt.figure(figsize=(10,6))
sns.heatmap(sales_pct,annot=True,fmt=".1f",cmap="YlOrRd",linewidths=0.5) #sns =Seaborn library(advanced visualization),annot=true=show numbers in boxes.
                                                                          #heatmap=Convert table to colored grid, fmt=".1f"=.decimal,1place,float. 
                                                                          #cmap="YlOrRd"= cmap-choose color.low=yellow,mid=orange,high=red.
plt.title("Percentage of Sales by Pizza Category and Size")
plt.ylabel("Pizza Category")
plt.xlabel("Pizza Size")

plt.show()
```

#### Total Pizza Sold by Pizza Category

```python
pizzas_by_category = df.groupby('pizza_category')['quantity'].sum() #df.groupby('pizza_category')=Group all rows based on categories like Veg,Non-Veg.
                                                                    #['quantity']=Selects only the quantity column.
                                                                    #sum()=Adds total quantity for each category
ax = pizzas_by_category.plot(     #.plot()=Show my data in a graph
    kind='bar',
    figsize=(8,5),
    color='red',
    edgecolor='black'
)

plt.title("Total Pizzas Sold by Pizza Category")
plt.xlabel("Pizza Category")
plt.ylabel("Total Pizzas Sold")
plt.xticks(rotation=0)

for i, val in enumerate(pizzas_by_category):     #here i is x axis and val is actual value,enumerate is returning both index and eliment at same time.
    plt.text(i, val + 5, str(val), ha='center', va='bottom',   #val = height of the bar,,+ 5 = move text a little above the bar.
             fontsize=9, fontweight='bold')                    #If bar height = 1500 than Text position would be 1505.

plt.tight_layout()
plt.show()
```

#### Top 5 Best Selling Pizza-Quantity Wise
